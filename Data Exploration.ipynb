```python
import matplotlib.pyplot as plt
import os
import numpy as np
from PIL import Image
```

```python
# Load 1 image from the test set
path1 = r'C:\Users\jandr\Github Repos\aBennecke\CS226_Final_Project\chestXray1.jpeg'
path2 = r'C:\Users\jandr\Github Repos\aBennecke\CS226_Final_Project\chestXray2.jpeg'

xray1 = plt.imread(path1)
xray2 = plt.imread(path2)

print(f'xray1 shape: {xray1.shape}')
print(f'xray2 shape: {xray2.shape}')

plt.imshow(xray1);
```

```python
# Load test data
testImages = []
path = r'C:\Users\jandr\Github Repos\aBennecke\CS226_Final_Project\chest_xray\test\NORMAL'
for filename in os.listdir(path):
    imgpath = path+"\\"+filename
    img = Image.open(imgpath)
    testImages.append(img)
```

```python
# Get average image width and height for resizing
allWidth  = []
allHeight = []

for img in testImages:
    allWidth.append(img.size[0])
    allHeight.append(img.size[1])
    
avgWidth = int(sum(allWidth)/len(allWidth))
avgHeight = int(sum(allHeight)/len(allHeight))

print(f'Average Image Size: ({avgWidth},{avgHeight})')
```

```python
# Histograms of width and height
plt.subplot(121)
plt.hist(allWidth); plt.title("Width of test images");
plt.subplot(122)
plt.hist(allHeight); plt.title("Height of test images");
```

```python
# Resizes all images to the mean of the test set
testResized = []
newSize = (avgWidth,avgHeight)
for img in testImages:
    testResized.append(img.resize(newSize))
```

```python
# Determine range of pixel values in a given image

allPixels = np.asarray(testResized[0])
allPixelsArr = np.array(allPixels)
np.unique(allPixelsArr)
```
